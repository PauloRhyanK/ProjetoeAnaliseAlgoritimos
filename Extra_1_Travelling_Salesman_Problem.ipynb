Gerar a Matriz de Caminhos:

```
def matrizCaminhos(N = 5):
  '''
    DOCSTRING:
    Gerar a Matriz de Caminhos:
    CAMINHOS:
    CAMINHO 1 : C1 - C2 - C3 - C4 - C5 - C1
    CAMINHO 2 : C1 - C2 - C3 - C5 - C4 - C1
    CAMINHO 3 : C1 - C2 - C4 - C3 - C5 - C1
    CAMINHO 4 : C1 - C2 - C4 - C5 - C3 - C1
    CAMINHO 5 : C1 - C2 - C5 - C3 - C4 - C1
    CAMINHO 6 : C1 - C2 - C5 - C4 - C3 - C1

    CAMINHO 7 : C1 - C3 - C2 - C4 - C5 - C1
    CAMINHO 8 : C1 - C3 - C2 - C5 - C4 - C1
    CAMINHO 9 : C1 - C3 - C4 - C2 - C5 - C1
    CAMINHO 10: C1 - C3 - C4 - C5 - C2 - C1
    CAMINHO 11: C1 - C3 - C5 - C2 - C4 - C1
    CAMINHO 12: C1 - C3 - C5 - C4 - C2 - C1

    CAMINHO 13: C1 - C4 - C2 - C3 - C5 - C1
    CAMINHO 14: C1 - C4 - C2 - C5 - C3 - C1
    CAMINHO 15: C1 - C4 - C3 - C2 - C5 - C1
    CAMINHO 16: C1 - C4 - C3 - C5 - C2 - C1
    CAMINHO 17: C1 - C4 - C5 - C2 - C3 - C1
    CAMINHO 18: C1 - C4 - C5 - C3 - C2 - C1

    CAMINHO 19: C1 - C5 - C2 - C3 - C4 - C1
    CAMINHO 20: C1 - C5 - C2 - C4 - C3 - C1
    CAMINHO 21: C1 - C5 - C3 - C4 - C5 - C1
    CAMINHO 22: C1 - C5 - C3 - C5 - C4 - C1
    CAMINHO 23: C1 - C5 - C4 - C2 - C3 - C1
    CAMINHO 24: C1 - C5 - C4 - C3 - C2 - C1

    Há rotas iguais? Quantas ?

    CAMINHO 1 = CAMINHO 24:

  '''
  i = 1
  for perm in permutations(range(1, N)):    # começando da cidade 1: 0 (index)
    print(f'CAMINHO {i}: {perm}')
    i += 1
  return
```

## Letra B)

Gerar todas as soluções: CAMINHOS + CUSTOS e apontar a solução ótima por força bruta.

```
def solucaoOtima(N=5, matriz_custo=None):
    '''
    DOCSTRING:
    Gera todas as soluções do problema do Caixeiro Viajante por força bruta,
    calcula os custos e aponta a solução ótima (menor custo).
    '''
    if matriz_custo is None:
        raise ValueError("A matriz de custo deve ser fornecida!")

    if len(matriz_custo) != N or any(len(row) != N for row in matriz_custo):
        raise ValueError("A matriz de custo deve ser NxN.")

    caminhos = list(permutations(range(1, N)))  # Gera todas as permutações
    custos = []

    for caminho in caminhos:
        custo = 0
        caminho_completo = (0,) + caminho + (0,)  # Inclui a volta ao ponto inicial
        for i in range(N):
            custo += matriz_custo[caminho_completo[i]][caminho_completo[i + 1]]
        custos.append(custo)

    # Determina o caminho ótimo (menor custo)
    menor_custo = min(custos)
    indice_melhor = custos.index(menor_custo)
    melhor_caminho = (0,) + caminhos[indice_melhor] + (0,)

    print("Caminhos gerados e seus custos:")
    for i, (caminho, custo) in enumerate(zip(caminhos, custos), start=1):
        caminho_formatado = " - ".join(f"C{c + 1}" for c in (0,) + caminho + (0,))
        print(f"CAMINHO {i}: {caminho_formatado} | CUSTO: {custo}")

    print("\nSolução Ótima:")
    caminho_otimo_formatado = " - ".join(f"C{c + 1}" for c in melhor_caminho)
    print(f"CAMINHO ÓTIMO: {caminho_otimo_formatado} | CUSTO MÍNIMO: {menor_custo}")

    return custos, melhor_caminho
```

```
# Chamada da função
matriz = matrizCusto()

solucaoOtima(N=5, matriz_custo=matriz)
```

## Letra C)

Reduzir a estrutura da Matriz de Custo para o valor mínimo de tamanho.

```
# Qual a quantidade mínima de espaço que devem ser alocado para essa instância?
def matrizCustoOtima(N=5):
    """
    Reduz a matriz de custo ao tamanho mínimo necessário, armazenando apenas
    a metade inferior (excluindo a diagonal principal) em um vetor 1D.
    """
    tamanho = FN(N)  # Utilizando FN(N) como custo de preenchimento
    vetorCusto = np.zeros(tamanho, dtype=int)

    # Criando um índice para armazenar na estrutura reduzida
    index = 0
    matrizCompleta = matrizCusto()  # Obtendo a matriz completa de custos

    # Preenchendo o vetor com a metade inferior da matriz de custos
    for i in range(N):
        for j in range(i + 1, N):  # Percorre apenas a parte triangular superior
            vetorCusto[index] = matrizCompleta[i][j]
            index += 1

    return vetorCusto
```

```
def solucaoOtima(vetorCusto, N=5):
    """
    Encontra a solução ótima do TSP com base no vetor de custos otimizado e imprime.
    """
    # Recriar a matriz de custos completa a partir do vetor otimizado
    matrizCustoCompleta = np.zeros((N, N), dtype=int)
    index = 0
    for i in range(N):
        for j in range(i + 1, N):
            matrizCustoCompleta[i][j] = vetorCusto[index]
            matrizCustoCompleta[j][i] = vetorCusto[index]  # Matriz simétrica
            index += 1

    # Gerar todas as permutações possíveis de caminhos
    melhoresCaminho = None
    melhorCusto = float('inf')

    for perm in permutations(range(N)):
        custo = 0
        # Calcula o custo do caminho atual
        for i in range(N - 1):
            custo += matrizCustoCompleta[perm[i]][perm[i + 1]]
        custo += matrizCustoCompleta[perm[-1]][perm[0]]  # Retorno à cidade inicial

        # Atualiza a melhor solução encontrada
        if custo < melhorCusto:
            melhoresCaminho = perm
            melhorCusto = custo

    # Imprime o resultado
    print(f'Melhor Caminho: {melhoresCaminho}')
    print(f'Custo Total: {melhorCusto}')
```

```
# Testando a função
vetorCusto = matrizCustoOtima(N=5)
solucaoOtima(vetorCusto, N=5)
```

## Letra E)

Reduzir a matriz de caminhos para o valor mínimo de tamanho.

```
def matrizCaminhosOtima(N=5):
    """
    Reduz a matriz de caminhos para o valor mínimo de tamanho necessário,
    eliminando os caminhos espelhados.
    A quantidade mínima de caminhos é dada por (N-1)! // 2.
    """
    # Calcula o número mínimo de caminhos a serem testados
    numCaminhos = math.factorial(N - 1) // 2  # Eliminar caminhos espelhados
    matrizOtima = np.empty((numCaminhos, N-1), dtype=int)  # Estrutura otimizada para armazenar os caminhos

    caminhos = list(permutations(range(1, N)))  # Gera todas as permutações possíveis (sem considerar cidade fixa)

    # Filtrando os caminhos espelhados (caminho e sua inversão)
    index = 0
    for i in range(len(caminhos)):
        caminho = caminhos[i]
        caminho_invertido = tuple(reversed(caminho))

        # Verifica se o caminho invertido já foi incluído, se não, inclui o caminho original
        if caminho_invertido not in caminhos[:i]:
            matrizOtima[index] = caminho
            index += 1

    return matrizOtima
```

```

# Testando a função
N = 5
matrizOtima = matrizCaminhosOtima(N)
print(f'Matriz de Caminhos Otimizada (Sem Caminhos Espelhados): \n{matrizOtima}')
```
