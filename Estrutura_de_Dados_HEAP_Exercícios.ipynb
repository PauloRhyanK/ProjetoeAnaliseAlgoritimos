## HEAP MÁXIMO

image.png

image.png

```
def TROCAR (array, i, iMaior):
  array[i], array[iMaior] = array[iMaior], array[i]
```

```
def MAX_HEAPIFY(array, i):
  l = LEFT(i)
  r = RIGHT(i)
  TAM = len(array) - 1
  if(l <= TAM and array[l] > array[i]):
    indexMaior = l;
  else:
    indexMaior = i;
  if(r <= TAM and array[r] > array[indexMaior]):
    indexMaior = r;
  if(indexMaior != i):
    TROCAR(array, i, indexMaior)
    MAX_HEAPIFY(array, indexMaior)
```

## CONSTRUIR A HEAP

<center>

image.png

```
def BUILD_MAX_HEAP (array):
  TAMANHO = len(array)
  for i in range(TAMANHO // 2 - 1, -1, -1):
    MAX_HEAPIFY(array, i)
```

# HEAPSORT:

**Propriedadades da Função:**

1.  A repetição começa com a ÚLTIMA FOLHA da HEAP: Variação $i = [(TAMANHO ARRAY - 1) ... 1]$;

1.  Chamar a função TROCAR: RAIZ com a ÚLTIMA FOLHA da HEAP;

1.  A função recursiva MAX HEAPPIFY mantém a propriedade: HEAP MÁXIMO.


image.png

Faça a implementação da função DEF HEAPSORT:

```
array = np.random.randint(10, 100, 10)
array
```

```
def heapsort(array):
  """
  DOCSTRING:
  Heap (“monte”): é uma Estrutura de Dados do tipo Objeto Arranjo Vetor
  que pode ser visto como uma árvore binária praticamente completa.
  Um Heap pode ser configurado como Heap Mínimo ou Heap Máximo.
  """
  BUILD_MAX_HEAP(array)        # f(n) = n
  TAMANHO = len(array) - 1
  for i in range (TAMANHO, 0, -1):
    TROCAR(array, 0, i)
    TamanhoHeap = i;                     # diminuir o tamanho da heap em 1
    MAX_HEAPIFY(array[:TamanhoHeap], 0)  # retira a última folha
  return None
```
